# Visibility Regions Merge

Divide-and-conquer algorithm for finding visibility regions.

```python
import math
import sympy as sp
import matplotlib.pyplot as plt
import matplotlib.patches as mplp
import matplotlib.collections as mplc

from stuff import *
```

We use a sympy matrix (aliased as `Vec2`) of two numbers to represent a point and a named tuple (called `Segment`) with two points (`a` and `b`) to represent a line segment (see file [`stuff.py`](stuff.py)).

The point from where we will be viewing will be the origin.

To represent a visibility region we use a list of 4-tuples. Each 4-tuple, named `VRay`, has a unit vector representing a direction, its angle with relation to the positive X axis and two scalars explained later.

```python
origin = Vec2([0, 0])

VRay = namedtuple('VRay', 'a v r l')
```

So for example we can have the following segments and their visibility region from the origin.

```python
segments = [Segment(Vec2([7, 9]), Vec2([-5, 2])),
            Segment(Vec2([-7, -7]), Vec2([0, -7])),
            Segment(Vec2([5, -5]), Vec2([10, 0]))]

vregion = []

for s in segments:
    vregion.append(VRay(atan2(s.a), normalize(s.a), -1, norm(s.a)))
    vregion.append(VRay(atan2(s.b), normalize(s.b), norm(s.b), -1))
```

To draw the segments and visibility region we use this `draw` function.

```python
def draw(segments=[], region=[], xlim=12, ylim=12, figsize=8):

    rlim = math.ceil(max(xlim, ylim) * math.sqrt(2.0))

    fig, ax = plt.subplots(figsize=(figsize, figsize))

    # Draw line segments

    lc = mplc.LineCollection(segments, linewidths=2)
    ax.add_collection(lc)

    # Draw triangles and wedges

    patches = []
    unbounded_regions = []

    iv = Vec2([1, 0])

    for i, vec in enumerate(region):

        pvec = region[i - 1] if i > 0 else VRay(0, iv, -1, -1)

        if vec.r < 0 and vec.v != pvec.v:
            unbounded_regions.append((float(pvec.a), float(vec.a)))

        if vec.r > 0 and pvec.l > 0 and vec.v != pvec.v:
            patches.append(mplp.Polygon([origin, pvec.v * pvec.l, vec.v * vec.r], True))

    if len(region) > 0 and region[-1].v != iv:
        unbounded_regions.append((float(region[-1].a), 0))

    for (t0, t1) in unbounded_regions:
        patches.append(mplp.Wedge([0, 0], rlim, t0 * 180 / math.pi, t1 * 180 / math.pi))

    pc = mplc.PatchCollection(patches, alpha=0.4)
    ax.add_collection(pc)

    # Draw origin

    ax.plot([origin], marker='o')

    # mpl stuff

    ax.margins(0.1)
    ax.set_aspect('equal', 'box')

    ax.set_xlim(-xlim, xlim)
    ax.set_ylim(-ylim, ylim)

    plt.show(block=True)
```

The previous segments and their visibility region would be drawn like this.

```python
draw(segments, vregion)
```

## Algorithm

The algorithm is inspired by mergesort in which we divide the unsorted list into `n` sublists, each containing 1 element, then we merge the them together until there is only one list, the final sorted list.

The case of visibility regions is similar, find the visibility region for each segment and then merge them together to find the final visibility region.

The visibility region for a segment would be a list of 2 `VRay`s each for each point of the segment. The scalar values accompaning the vector are the lenght of how far can you see on the right and on the left side of the vector. We use a negative value to represent infinity.

```python
def vrays_for_segment(p):
    a = p.a
    b = p.b

    if cross(a, b) < 0:
        a, b = b, a

    atan2b = atan2(b) if atan2(b) != 0 else 2 * sp.pi

    return [VRay(atan2(a), normalize(a), -1, norm(a)), VRay(atan2b, normalize(b), norm(b), -1)]
```

Now, there is an exception to this when the segment crosses the positive X axis. In that case we split the segment into two segments.

```python
def visibility_region_for_segment(p):
    a = p.a
    b = p.b

    if cross(a, b) < 0:
        a, b = b, a

    s = b - a

    if s[1] == 0:
        return vrays_for_segment(p)

    i = -a[1] / s[1]

    if 0 < i < 1 and a[0] + s[0] * i > 0:
        c = a + s * i

        return vrays_for_segment(Segment(c, b)) + vrays_for_segment(Segment(a, c))

    return vrays_for_segment(p)
```

We also use other auxiliary functions:

- `ray_segment_intersection` finds the intersection of ray `v` with segment `q`, if any.
- `cut_segment` cuts the segment `s` by the rays `v` and `w`.

```python
def ray_segment_intersection(p, v):
    u = p.b - p.a

    w = p.a

    s = cross(u, v)

    if s == 0:
        return None

    r = w / s
    i = cross(v, r)

    if 0 <= i <= 1:
        return p[0] + u * i

    return None


def cut_segment(s, v, w):
    a = ray_segment_intersection(s, v)
    b = ray_segment_intersection(s, w)

    if a is None or b is None:
        return None

    return Segment(a, b)
```

The merge function sorts the `VRay` lists by their angle while updating how far can they see. 

```python
def merge(a, b):
    c = [a, b]
    n = [len(a), len(b)]
    i = [0, 0]

    s = []

    while i[0] < n[0] or i[1] < n[1]:

        if i[0] == n[0] or i[1] == n[1]:
            sel = 0 if i[0] < n[0] else 1
        else:
            sel = 0 if c[0][i[0]].a < c[1][i[1]].a else 1

        s.append(c[sel][i[sel]])

        if len(s) == 1:
            i[sel] = i[sel] + 1
            continue

        p = []

        for j in range(2):
            if 0 < i[j] < n[j] and c[j][i[j]].r > 0 and c[j][i[j] - 1].l > 0:
                v = c[j][i[j]]
                w = c[j][i[j] - 1]
                p.append(Segment(w.v * w.l, v.v * v.r))

        p = sorted(filter(None, [cut_segment(q, s[-2].v, s[-1].v) for q in p]), key=lambda q: tuple(map(norm, q)))

        if p:
            s[-2] = s[-2]._replace(l=norm(p[0][0]))
            s[-1] = s[-1]._replace(r=norm(p[0][1]))

        i[sel] = i[sel] + 1

    return s
```

Then we put it all together like in merge sort.

```python
def visimerge(s):
    n = len(s)
    m = n // 2

    if n == 0:
        return []

    if n == 1:
        return visibility_region_for_segment(s[0])

    s0 = visimerge(s[0:m])
    s1 = visimerge(s[m:n])

    return merge(s0, s1)
```

Before calling `visimerge` we filter out collinear segments with the origin since these segments don't interfere with the visibility.

```python
def visibility_region(s):
    r = [p for p in s if not are_collinear(origin, *p)]
    return visimerge(r)
```

### Test

```python
with open('datasets/set0.txt') as f:
    segments = parse(f.read())

xlim = 0
ylim = 0

for s in segments:
    xlim = max(xlim, abs(s[0][0]), abs(s[1][0]))
    ylim = max(ylim, abs(s[0][1]), abs(s[1][1]))

xlim = float(xlim) * 1.05
ylim = float(ylim) * 1.05

vis = visibility_region(segments)

draw(segments, vis, xlim=xlim, ylim=ylim, figsize=10)
```
